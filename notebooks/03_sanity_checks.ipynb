# Notebook 03 — Quick Sanity Checks

Small checks to make sure the dataset loads and basic columns exist.

```python
import pandas as pd

df = pd.read_csv('../data/sample/diabetes_sample.csv')

# basic checks
assert 'Diabetes_012' in df.columns
assert df.shape[0] > 0
assert df.isna().mean().max() < 0.5  # no column should be >50% missing in this sample

print('✅ basic checks passed')
```

```python
# check that the binary conversion works
at_risk = (df['Diabetes_012'] != 0).astype(int)
print(at_risk.value_counts())
```
