# Notebook 01 — Quick EDA (Diabetes Sample)

This notebook does a quick beginner-friendly EDA on the sample dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/sample/diabetes_sample.csv')
df.head()
```

```python
df.shape
```

```python
df['Diabetes_012'].value_counts().sort_index()
```

```python
# Plot original 3-class distribution
counts = df['Diabetes_012'].value_counts().sort_index()
plt.figure()
counts.plot(kind='bar')
plt.title('Target distribution (0=healthy,1=prediabetes,2=diabetes)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Create binary label: 0=healthy, 1=at risk

df['at_risk'] = (df['Diabetes_012'] != 0).astype(int)
df['at_risk'].value_counts()
```

```python
# Basic missing values check

df.isna().sum().sort_values(ascending=False).head(15)
```

```python
# Simple summary stats

df.drop(columns=['Diabetes_012']).describe().T.head(10)
```
